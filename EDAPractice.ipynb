```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic=pd.read_excel("E:\\Datasets\\tested.xlsx")
titanic.head()
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
iris=pd.read_csv('E:\\Datasets\\iris.csv')
iris.head()
```

# Scatter ploat--Numerical+Numerical

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

# bar plot--Numeriacal+Categorical

```python
sns.barplot(titanic["Pclass"],titanic['Age'])
```

# box plot--Numeriacal+Categorical

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

# Distplot plot--Numeriacal+Categorical

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# Hitmap plot--Categorical+Categorical

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

# Clustermap plot--Categorical+Categorical

```python
 data=pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(data)
```

# PairPlot

```python
sns.pairplot(iris,hue='species')
```
